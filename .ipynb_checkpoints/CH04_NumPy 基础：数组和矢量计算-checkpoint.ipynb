```python
temp=arr[3:7]
#注意这波操作，会改变原数组
temp[:]=123456
arr
```

### 利用数组进行数据处理
用数组表达式代替循环的做法，通常被称为矢量化。一般来说，矢量化
数组运算要比等价的纯Python方式快上一两个数量级（甚至更多） 
作为简单的例子，假设我们想要在一组值（网格型） 上计算函
数 sqrt(x^2+y^2) 。np.meshgrid函数接受两个一维数组，并产生两个二维矩阵
（对应于两个数组中所有的(x,y)对） ：

```python
#1000个点
points = np.arange(-5, 5, 0.01)
xs, ys = np.meshgrid(points, points)
xs,ys
```

```python
xs.shape
```

```python
z = np.sqrt(xs ** 2 + ys ** 2)
z,z.shape
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(xs,cmap=plt.cm.gray); plt.colorbar()
```

```python
plt.imshow(ys, cmap=plt.cm.gray); plt.colorbar()
```

```python
plt.title("Image plot of $\sqrt{x^2 + y^2}$ for a gridof values")
plt.imshow(z, cmap=plt.cm.gray); plt.colorbar()
```

* 将条件逻辑表述为数组运算
numpy.where函数是三元表达式x if condition else y的矢量化版本

比纯python的效率更高，并且可以用到高维度的数组上

```python
xarr = np.array([1.1, 1.2, 1.3, 1.4, 1.5])
yarr = np.array([2.1, 2.2, 2.3, 2.4, 2.5])
cond = np.array([True, False, True, True, False])
```

```python
#假设我们想要根据cond中的值选取xarr和yarr的值，当cond的值为真则x否则y
result= np.where (cond,xarr,yarr)
result
```

```python
arr=np.random.randn(5,5)
arr
```

```python
#后两个参数不比是数组，也可以是数值，这样就替换了原数组
np.where(arr>0,555,888)
```

```python
np.where(arr>0,arr,0)
```

### 4.7 示例：随机漫步
从0开始，步长1和－1出现的概率相等。

```python
import random
p=0
walk=[p]
steps=1000
for i in range(steps):
    # step= （1 if random.randint(0,1) else -1）
    step= 1 if random.randint(0,1) else -1
    p+=step
    walk.append(p)
```

```python
plt.plot(walk)
```

```python
#Numpy的实现
draws=np.random.randint(0,2,steps)
step=np.where(draws>0,1,-1)
walk=step.cumsum()
```

```python
plt.plot(walk)
```

```python
walk.max()
```

```python
np.abs(walk) >= 10
```

```python
#argmax来解决这个问题，它返回的是该布尔型数组第一个最大值的索引（True就是最大值） ：
(np.abs(walk) >= 10).argmax()
```

* 一次模拟多个随机漫步
换成多维数组就可以


```python
nwalks = 5000
nsteps = 1000
draws = np.random.randint(0, 2, size=(nwalks, nsteps))
```

```python
steps = np.where(draws > 0, 1, -1)
walks = steps.cumsum(1)
walks
```

```python
walks.max()
```

```python
#统计超过30的穿越时间，首先看这个5000次有多少超过了30的
hits30=(np.abs(walks)>30).any(1)
hits30
```

```python
hits30.sum()
```

```python
walks[hits30]
```

```python
#选出大于30的数据,第一次出现的时候
crossing_times = (np.abs(walks[hits30]) >= 30).argmax(1)
crossing_times
```

```python
crossing_times.mean()
```
